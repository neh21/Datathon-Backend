```python
districts = ['Bagalkot', 'Ballari', 'Belagavi City', 'Belagavi Dist',
             'Bengaluru City', 'Bengaluru Dist', 'Bidar', 'Chamarajanagar',
             'Chickballapura', 'Chikkamagaluru', 'Chitradurga', 'CID',
             'Coastal Security Police', 'Dakshina Kannada', 'Davanagere',
             'Dharwad', 'Gadag', 'Hassan', 'Haveri', 'Hubballi Dharwad City',
             'ISD Bengaluru', 'K.G.F', 'Kalaburagi']
coordinates = [
    (16.18, 75.7),   # Bagalkot
    (15.15, 76.92),  # Ballari
    (15.87, 74.5),   # Belagavi City
    (15.87, 74.5),   # Belagavi Dist
    (12.97, 77.59),  # Bengaluru City
    (12.97, 77.59),  # Bengaluru Dist
    (17.91, 77.52),  # Bidar
    (11.92, 76.94),  # Chamarajanagar
    (13.43, 77.73),  # Chickballapura
    (13.32, 75.77),  # Chikkamagaluru
    (14.23, 76.4),   # Chitradurga
    (12.909367,77.638757), # Placeholder for CID
    (13.087983, 74.773067),# Placeholder for Coastal Security Police
    (12.9, 74.85),   # Dakshina Kannada
    (14.46, 75.92),  # Davanagere
    (15.46, 75.01),  # Dharwad
    (15.43, 75.63),  # Gadag
    (13.01, 76.1),   # Hassan
    (14.79, 75.4),   # Haveri
    (15.46, 75.01),  # Hubballi Dharwad City
    (12.968800, 77.616000),# Placeholder for ISD Bengaluru
    (12.97, 77.59),  # K.G.F
    (17.33, 76.83),  # Kalaburagi
]
districts_coordinates = dict(zip(districts, coordinates))
```

```python
for district, coords in districts_coordinates.items():
    df.loc[(df['District_Name'] == district) & (df['Latitude'] == 0.0), 'Latitude'] = coords[0]
    df.loc[(df['District_Name'] == district) & (df['Longitude'] == 0.0), 'Longitude'] = coords[1]
```
