```python
import numpy as np

for unit in df['UnitName'].unique():
    # Filter to get non-zero latitudes and longitudes for the unit
    valid_latitudes = df[(df['UnitName'] == unit) & (df['Latitude'] != 0.0)]['Latitude']
    valid_longitudes = df[(df['UnitName'] == unit) & (df['Longitude'] != 0.0)]['Longitude']
    print(unit)
    if not valid_latitudes.empty and not valid_longitudes.empty:
        # Replace each zero value with a unique random value within the observed range for the unit
        for index, row in df[(df['UnitName'] == unit) & ((df['Latitude'] == 0.0) | (df['Longitude'] == 0.0))].iterrows():
            if row['Latitude'] == 0.0:
                df.at[index, 'Latitude'] = np.random.uniform(valid_latitudes.min(), valid_latitudes.max())
            if row['Longitude'] == 0.0:
                df.at[index, 'Longitude'] = np.random.uniform(valid_longitudes.min(), valid_longitudes.max())
```

```python
import pandas as pd
import numpy as np

# Assume df is your existing DataFrame
i = 0
def process_distance(direction_str):
    import re
    global i
    # Convert any non-string input to string
    direction_str = str(direction_str)

    # Find and extract the numeric distance
    distance_match = re.search(r'\d+', direction_str)
    distance = int(distance_match.group()) if distance_match else np.nan

    # Standardize and extract the direction
    direction_str = direction_str.upper()
    if 'EAST' in direction_str:
        direction = 'EAST'
    elif 'WEST' in direction_str:
        direction = 'WEST'
    elif 'SOUTH' in direction_str:
        direction = 'SOUTH'
    elif 'NORTH' in direction_str:
        direction = 'NORTH' 
    else:
        direction = np.nan  # Handle the case where direction is not specified
    i+=1
    if (i%10000==0):
        print(i)
    return pd.Series([distance, direction], index=['Distance', 'Direction'])

# Apply the function to the 'Distance from PS' column
df[['Distance', 'Direction']] = df['Distance from PS'].apply(process_distance)

# Now, to fill null values
# # Fill 'Distance' with the median value (or mean, depending on your data distribution and preference)
# df['Distance'].fillna(df['Distance'].median(), inplace=True)

# # Fill 'Direction' with the mode
# df['Direction'].fillna(df['Direction'].mode()[0], inplace=True)
```

```python
# Fill null values in 'Distance' with the median value of the column
df['Distance'].fillna(df['Distance'].median(), inplace=True)

# Fill null values in 'Direction' with the most frequent value ('mode') of the column
df['Direction'].fillna(df['Direction'].mode()[0], inplace=True)
```
